## Задание 4
Есть 4 кубика. На трех из них окрашена белым половина граней, а на четвертом кубике всего одна грань из шести белая. Наудачу выбранный кубик подбрасывается семь раз. Найти вероятность того, что был выбран четвертый кубик, если при семи подбрасываниях белая грань выпала ровно один раз. Ответ округлите до тысячных.

```python
import pandas as pd
from scipy import special
from math import factorial
```

```python
# найдем вероятность того, что белая грань выпадет при броске одного из первых 3х кубиков (с половиной белых граней) по формуле Бернулли
def white1 (n, k):
    p, q = 3/6, 3/6
    C = factorial(n)/(factorial(k)*factorial(n-k))
    return round(C*p**k*q**(n-k), 4)

white1(7, 1)
```

```python
# найдем вероятность того, что белая грань выпадет при броске 4того кубиков (с 1 из 6 белых граней) по формуле Бернулли
def white2 (n, k):
    p, q = 1/6, 5/6
    C = factorial(n)/(factorial(k)*factorial(n-k))
    return round(C*p**k*q**(n-k), 4)

white2(7, 1)
```

```python
# тогда по формуле полной вероятности получим вероятность выпадения белой грани (всего 4 кубика, 3 из них с половиной белых граней, один - с одной белой гранью, поэтому 3/4 и 1/4):
3/4 * white1(7, 1) + 1/4 * white2(7, 1)
```

```python
# оценим вероятность выпадания четвертого кубика белой гранью по формуле Байеса:
round(1/4 * white2(7, 1) / (3/4 * white1(7, 1) + 1/4 * white2(7, 1)), 4)
```
