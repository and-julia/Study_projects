<details>
<summary>&#9658; Нажмите сюда, чтобы увидеть подсказку</summary>
  <p>

К элементам кортежа (tuple) можно обращаться по индексам.

Взять элемент с индексом `2` из кортежа `my_tuple` можно так:

```python
my_tuple = (36.6, 39.0, 37.1, 36.8)
my_tuple[2]
```

**ВАЖНО!** Индексация в Python начинается с 0!

Чтобы задать в переменную элемент из кортежа с определённым индексом, нужно обратиться к этому элементу по индексу и присвоить полученное значение новой переменной.

```python
element = my_tuple[2]
```

Кроме того, в Python можно делать распаковку кортежей. Это удобный способ извлечения значений из кортежа и присвоения их переменным.

Для этого в левой части выражения перечисляются переменные, а в правой — указывается кортеж:
      
```python
my_tuple = (1, 2, 3)
a, b, c = my_tuple
```
      
**Примечание:** если количество переменных не совпадает с количеством элементов в кортеже, Python выдаст ошибку.
      
      
</p>

Для начала с помощью функции `range()` сгенерируйте последовательность значений от `120` до `240` с шагом `5`. Подумайте, какое значение нужно указать у параметра `stop`, чтобы значение `240` тоже оказалось включённым в последовательность.

  <p>

Для генерации последовательности необходимо использовать функцию `range()`:

```python
range(start, stop, step)
```
Такая запись создаёт набор целых чисел от `start` до `stop` с шагом `step`. По умолчанию `start` равен 0, а `step` равен 1. При этом конечное значение `stop` в результат не включается.
      
Чтобы значение, равное `stop`, оказалось включенным в последовательность, необходимо указать конечное значение на 1 больше.
      
Сгенерированная последовельность не является списком, поэтому для дальнейших расчётов необходимо сконвертировать её в список с помощью функции `list()`:
      
```python
list(range(start, stop, step))
```
**Не забудьте** сохранить список в переменную!      

</p>
</details>
